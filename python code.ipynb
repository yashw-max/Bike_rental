```python
# clearly we can see that outliers are present in casual, windspeed, hum variables 
#Detect & Delete Outliers
pnames = ["casual"]

for i in pnames :
    print (i)
    q75,q25 = np.percentile(df.loc[:,i],[75,25])
    iqr = q75-q25
    
    min = q25 - (iqr*1.5)
    max = q75 + (iqr*1.5)
    print (min)
    print (max)
    
    df = df.drop(df[df.loc[:,i] < min].index)
    df = df.drop(df[df.loc[:,i] > max].index)
```

```python
############# Model selection #####################
#diividing  Test and train data  using skilearn   train_test_split 
from sklearn.model_selection import train_test_split
```

```python
df.columns
```

```python

X =df[['season', 'yr', 'mnth', 'holiday', 'weekday', 'workingday',
       'weathersit', 'temp', 'windspeed', 'casual', 'registered']]
y = df['cnt']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)

#X_train.shape 
#y_train.shape
```

```python
#********************************** Decision Tree  Regressor ************************************
#Importing Decision Tree Regressor from sklear.tree
from sklearn.tree import DecisionTreeRegressor
```

```python
dtree = DecisionTreeRegressor()
```

```python
dtree.fit(X_train,y_train)
```

```python
## predition   by using predict method
predictions_DT = dtree.predict(X_test)
#print(predictions_DT)
```

```python
sns.distplot((y_test-predictions_DT))
# residuals are normally distributed as we can say that model as predicted good. 
```

```python
 def MAPE(y_true, y_pred): 
    mape = np.mean(np.abs((y_true - y_pred) / y_true))*100
    return mape

MAPE(y_test, predictions_DT)

# error = 3.96
```

```python
#Control overfitting by setting "max_depth" to 10 and "min_samples_split" to 5
max_depth = 10
min_samples_split =5
dtree = DecisionTreeRegressor(max_depth =max_depth , min_samples_split =min_samples_split, random_state = 1)
dtree.fit(X_train,y_train)
print(dtree)

predictions_DT = dtree.predict(X_test)

MAPE(y_test,predictions_DT)

```

```python
#RSS - Residual Sum 
#test_target_feature,predictions_DT_two
rss= ((y_test-predictions_DT)**2).sum()
print(rss)

MSE = np.mean((y_test-predictions_DT)**2)
print(MSE)

#RMSE
rmse=np.sqrt(MSE)
print(rmse)


##MAPE = 3.96
### RMSE = 185.859
```

```python
#############Random Forest ######################
from sklearn.ensemble import RandomForestRegressor
rfc = RandomForestRegressor(n_estimators= 200, random_state=101)
rfc.fit(X_train,y_train)
```

```python
rfc_pred = rfc.predict(X_test)
#print(predictions_DT)
```

```python
#Evaluate Random forest using  MAPE 

MAPE(y_test,rfc_pred)

#Error rate is 1.76

#Here it is stating accuracy of the model increases
```

```python
#RSS - Residual Sum 
#test_target_feature,predictions_DT_two
rss= ((y_test-rfc_pred)**2).sum()
print(rss)

MSE = np.mean((y_test-rfc_pred)**2)
print(MSE)

#RMSE
rmse=np.sqrt(MSE)
print(rmse)

##MAPE = 1.76
### RMSE = 89.218
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Make predictions using the testing set
y_pred = lm.predict(X_test)
```
